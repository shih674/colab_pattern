```
!pip install beautifulsoup4
import requests
from bs4 import BeautifulSoup
```

# 取得RESPONSE

```
url = 'https://24h.pchome.com.tw/'

print(f'正在爬取網址: {url}')
response = requests.get(url).text
response
```

# 使用BeautifulSoup解析原始碼(response)


```
sourceCode = BeautifulSoup(response,'html.parser')
sourceCode
```

# 把原始碼保存成txt，方便搜尋(crtl+F)關鍵字


```
with open('sourceCode.txt', 'w', encoding='utf8') as f:
    f.write(str(sourceCode))
```
